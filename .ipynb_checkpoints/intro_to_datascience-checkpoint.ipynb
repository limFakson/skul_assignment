```python
# Import necessary libraries
from sklearn.metrics import f1_score, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier # example model
import pandas as pd
import numpy as np
```

```python
# Load your dataset
data = pd.read_csv('StudentsPerformance.csv')

# Prepare the features (X) and target (y)
x = data.drop(['avg score','gender','race/ethnicity','parental level of education','lunch','test preparation course'], axis=1)  # replace 'target_column' with your label column
y = data['avg score']
```

```python
# Spliting the data into test and train for the model build
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)

# covert the continuous value into int
y_train = np.round(y_train).astype(int)
y_test = np.round(y_test).astype(int)
```

```python
# Create a model from RandomForestClassifier function
model = RandomForestClassifier ()

# Train a model with the x and y train
model.fit(x_train, y_train)
```

```python
# Calcualting the train prediction accuracy
y_train_predict = model.predict(x_train)
train_accuracy = accuracy_score(y_train, y_train_predict)
train_accuracy
```

```python
# Calculate the test prediction accuracy 
y_test_predict = model.predict(x_test)
test_accuracy = accuracy_score(y_test, y_test_predict)
test_accuracy
```
