```python
# Import neccessary packages
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score, mean_squared_error, classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier # example model
import pandas as pd
import numpy as np
```

```python
# save train csv file in train data
train_data = pd.read_csv(r"C:\Users\USER\Desktop\Fakson_web\Programs\lessons\Python\Jupyter\dataset\vicsdata\plate-size.csv")
train_data.head()
```

```python
# split target and dataset
x = train_data.drop(['gender', 'Date', 'fasting'], axis=1)
y = train_data['gender'] # Target
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=100)
x_train.shape
```

```python
rl = RandomForestClassifier()
rl.fit(x_train, y_train)
```

```python
x_pred = rl.predict(x_train)
train_accuracy = accuracy_score(x_pred, y_train)
print(f'Train accuracy score: {train_accuracy}')
test_pred = rl.predict(x_test)
test_accuracy = accuracy_score(test_pred, y_test)
print(f'Test accuracy score: {test_accuracy}')
```

```python
report = classification_report(test_pred, y_test)
print(f'report{report}')
```
